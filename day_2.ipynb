### BMI

#### Calculate the BMI of a person with inputs on his height and weight

```python
height = input("enter your height in m: ")
weight = input("enter your weight in kg: ")
height = float(height)
weight = float(weight)
bmi = int(weight/(height**2))
print(bmi)
```

### Coding Challenge

#### Create a program using maths and f-Strings that tells us how many days, weeks, months we have left if we live until 90 years old. 

```python
age = input("What is your current age?")
age = int(age)
years = 90 - age
days = years * 365
weeks = years * 52
months = years * 12
print(f"You have {days} days, {weeks} weeks, and {months} months left")
```

### Tip Calculator

#### Design a tip calculator that show 1). How much to pay. 2). What percentage to pay? 3). How many people would pay

```python
print("Welcome to the tip calculator")
bill = input("What was the total bill?")
bill = float(bill)
percentage = input("What percentage tip would you like to give?")
percentage = int(percentage)
per_bill = bill*(percentage/100)
total_bill = per_bill + bill
people = input("How many people to split the bill with?")
people = int(people)
ind_bill = round(total_bill/people,2)
print(f"Every Person should pay {ind_bill}")
```
